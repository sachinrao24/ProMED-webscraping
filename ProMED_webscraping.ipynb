## Load dependencies

```python
import requests
import html5lib
from bs4 import BeautifulSoup
import re
from tqdm.auto import tqdm
import pandas as pd
```

## Getting all post IDs from ProMED

```python
url = 'https://promedmail.org/wp-admin/admin-ajax.php'
headers = {'user-agent':'Mozilla/5.0'}
```

### Page 1

```python
search_data_page1 = {'action': 'get_promed_search_content',
               'query[0][name]': 'kwby1',
               'query[0][value]': 'summary',
               'query[1][name]': 'search',
               'query[1][value]': '',
               'query[2][name]': 'date1',
               'query[2][value]':'02/20/2023',
               'query[3][name]': 'date2',
               'query[3][value]':'03/20/2023',
               'query[4][name]': 'feed_id',
               'query[4][value]': '1',
               'tz': 'Asia/Calcutta'}

response = requests.post(url, headers=headers, data=search_data_page1)
page1_content = response.json()
```

```python
page1_news = page1_content['results'].split("\n")

date_pattern = '<li> (.*) <a'
id_pattern = '.*id="id([0-9]+).*"'

post_dates_and_ids={}

i=0
for post in page1_news:
    y = re.search(date_pattern, page1_news[i])
    x = re.search(id_pattern, page1_news[i])
    post_dates_and_ids[x.group(1)]=y.group(1)
    i+=1
```

```python
post_dates_and_ids
```

### Pages 2 to 5

```python
for page_num in range(1,5):
    search_data = {'action': 'get_promed_search_content',
                         'query[0][name]': 'pagenum',
                         'query[0][value]': f'{page_num}',
                         'query[1][name]': 'kwby1',
                         'query[1][value]': 'summary',
                         'query[2][name]': 'search',
                         'query[2][value]': '',
                         'query[3][name]': 'date1',
                         'query[3][value]':'02/20/2023',
                         'query[4][name]': 'date2',
                         'query[4][value]':'03/20/2023',
                         'query[5][name]': 'feed_id',
                         'query[5][value]': '1',
                         'query[6][name]': 'timezone',
                         'query[6][value]': 'Asia/Calcutta',
                         'query[7][name]': 'submit',
                         'query[7][value]': 'next',
                         'tz': 'Asia/Calcutta'}

    response = requests.post(url, headers=headers, data=search_data)
    page_content = response.json()
    page_news = page_content['results'].split("\n")

    i=0
    for post in page_news:
        post_date = re.search(date_pattern, page_news[i])
        post_id = re.search(id_pattern, page_news[i])
        post_dates_and_ids[post_id.group(1)]=post_date.group(1)
        i+=1
```

```python
post_dates_and_ids
```

All post IDs and dates between 20-Feb-2023 and 20-March-2023 are now obtained.

## Getting individual post content

```python
source_pattern = 'Source:\s*(.*)\s.*\r'
no_source_pattern = '(?:Date:.*?\n)(.*?)\r'
post_ids = list(post_dates_and_ids.keys())
sources = []
texts = []
```

```python
progress_bar = tqdm(post_ids)
for post_id in progress_bar:
    progress_bar.set_description(post_id)
    search_post = {'action': 'get_latest_post_data',
                   'alertId': f'{post_id}',
                   'tz': 'Asia/Calcutta'}
    response = requests.post(url, headers=headers, data=search_post)
    try:
        post_content = response.json()
        if post_content['postinfo']!=None:
            post_text = post_content['postinfo']['content']
            texts.append(post_text)
            post_source = re.findall(source_pattern, post_text)
            if post_source!=None:
                unique_sources = set()
                if len(post_source)==0:
                    sources.append("No source")
                else:
                    for source in post_source:
                        unique_sources.add(source)
                    unique_sources_list = []
                    for source in unique_sources:
                        unique_sources_list.append(source)
                    sources.append(unique_sources_list)
            else:
                post_source = re.search(no_source_pattern, post_text)
                sources.append(post_source.group(1))
        else:
            sources.append("No source")
            texts.append("")
    except Exception as ex:
        print("Error: ",type(ex).__name__)
```

```python
index = range(0,236)
df = pd.DataFrame(post_dates_and_ids.keys(), index=index)
df.rename(columns={0:'post_id'}, inplace=True)
df['Date of article']=post_dates_and_ids.values()
df['Complete extracted text']=texts
df
```

## Cleaning raw post text

```python
post_text = df['Complete extracted text'][90]
post_text
```

```python
text_pattern = '(?:[\s\S]*)(?:Source:\s*.*\s.*\r)([\s\S]*)(?:\[See Also|\]\.\.\.)'
text_pattern_for_unspecified_source = '(?:[\s\S]*)(?:Date:\s*.*\s.*\r)([\s\S]*)(?:\[See Also|\]\.\.\.)'

extracted_text = re.search(text_pattern_for_unspecified_source, post_text)
print(extracted_text.group(1))
```

```python
progress_bar = tqdm(texts)
cleaned_list = []
for text in progress_bar:
    extracted_text = re.search(text_pattern, text)
    if extracted_text == None:
        extracted_text2 = re.search(text_pattern_for_unspecified_source, text)
        if extracted_text2 == None:
            cleaned_list.append('')
            continue
        else:
            extracted_text2 = extracted_text2.group(1)
    else:
        extracted_text2 = extracted_text.group(1)
    cleaned_text = re.sub('\r', '', extracted_text2)
    cleaned_text = re.sub('\n', '', cleaned_text)
    cleaned_list.append(cleaned_text)
```

```python
cleaned_list[:5]
```

```python
df['Cleaned text'] = cleaned_list
```

```python
clean_source_pattern = '(.*?)\s*\['
cleaned_sources = [] 
for i in range(0, len(sources)):
    if sources[i]=="No source":
        cleaned_sources.append("No source")
        continue
    temp = []
    for source in sources[i]:
        cleaned_source = re.search(clean_source_pattern, source)
        if cleaned_source==None:
            temp.append(source)
        else:
            temp.append(cleaned_source.group(1))
    sources_string=''
    for source_num in range(0, len(temp)-1):
        sources_string+=temp[source_num]+", "
    sources_string+=temp[-1]
    cleaned_sources.append(sources_string)
```

```python
df['Source of article (Website Name)'] = cleaned_sources
```
